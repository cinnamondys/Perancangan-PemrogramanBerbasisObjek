```python
import math

# x1, 2 = [(-b) +/- akar(b^2 -4*a*c)] / (2*a)

def rumusABC(a, b, c) :
    yield b**2 -4*a*c   # determinan
    yield 2*a    # nilai pembagi
    yield ((-b) + math.sqrt(b**2 - 4*a*c)) / (2*a)   # x1
    yield ((-b) - math.sqrt(b**2 - 4*a*c)) / (2*a)   # x2

a = rumusABC(2, 6, -56)
print(a)
print(f"Hasil Determinan : {next(a)}")
print(f"Nilai Pembagi    : {next(a)}")
print(f"Nilai x1         : {next(a)}")
print(f"Nilai x2         : {next(a)}")

b = rumusABC(0, 3, 7)
print(b)
print(f"Hasil Determinan : {next(b)}")
print(f"Nilai Pembagi    : {next(b)}")
print(f"Nilai x1         : {next(b)}")
print(f"Nilai x2         : {next(b)}")
```

```python
# Coroutine

def stokBarang() :
    jlh = 0 
    while True :
        brg = yield jlh
        jlh += brg

a = stokBarang()
print(f"Jumlah stok awal = {next(a)} barang")
print(f"Jumlah stok setelah ditambah 3 barang = {a.send(3)} barang")
print(f"Jumlah stok setelah ditambah 10 barang = {a.send(10)} barang")
print(f"Jumlah stok setelah dikurang 5 barang = {a.send(-5)} barang")
```

```python
# Case Study

class BUKU :
    perpus = []
    def inputBuku(self):
        kd = "Kode Kosong"
        nm = ""
        jns = ""
        cek = True
        while cek:
            kodeBuku = yield kd
            if(len(self.perpus) < 1):
                print("Kode Kosong")
                cek = False
            else:
                cek = False
                for i in range(len(self.perpus)):
                    if(self.perpus[i]['kode'] == kodeBuku):
                        cek = True
                if(cek):
                    print("Kode Buku sudah terdaftar ... !!!")
                    print("Masukkan kode buku lagi ... !!!")
        namaBuku = yield nm
        jenisBuku = yield jns
        tmp = {"kode" : kodeBuku, "nama" : namaBuku, "jenis" : jenisBuku}
        self.perpus.append(tmp)
        print(self.perpus)
    def cetakBuku(self):
        for i in range(len(self.perpus)):
            print(f"Buku ke-{i+1}:")
            print(f"Kode Buku  = {self.perpus[i]['kode']}")
            print(f"Nama Buku  = {self.perpus[i]['nama']}")
            print(f"Jenis Buku = {self.perpus[i]['jenis']}")
            print()

a = BUKU()
b = a.inputBuku()
print(b)
print(next(b))
try :
    b.send("IF001")
    b.send("Pemrograman Python")
    b.send("Pemrograman")
except StopIteration :
    print()

a.cetakBuku()

b = a.inputBuku()
print(b)
print(next(b))
try :
    b.send("IF002")
    b.send("Bahasa C")
    b.send("Pemrograman")
except StopIteration :
    print()

a.cetakBuku()
```

```python
# Latihan Task 1 :
 
class Mikroskil :
    mhs = []
    jlh = 0
    def tambahData(self) :
        noInduk = ""
        nama = ""
        noHp = ""
        cek = True
        while cek:
            NoInduk = yield noInduk
            if(len(self.mhs) < 1):
                cek = False
            else:
                cek = False
                for i in range(len(self.mhs)):
                    if(self.mhs[i]['NIM'] == NoInduk):
                        cek = True
                if(cek):
                    print("NIM sudah terdaftar ... !!!")
                    print("Masukkan NIM baru lagi ... !!!")
        Nama = yield nama
        NoHp = yield noHp
        tmp = {"NIM" : NoInduk, "Nama" : Nama, "NoHp" : NoHp}
        self.mhs.append(tmp)
        self.jlh += 1
    def cetakData(self) :
        mhs = self.mhs
        i = 1
        print("Data-data Mahasiswa :")
        for data in mhs :
            print(i, ".")
            print(f"NIM   : {data['NIM']}")
            print(f"Nama  : {data['Nama']}")
            print(f"No Hp : {data['NoHp']}")
            print()
            i+=1
    def absensi(self, noInduk) :
        mhs = self.mhs
        if len(mhs) == 0 :
            print("Data masih kosong, tidak ada yang bisa dicari")
        else :
            for posisi in range(len(mhs)) :
                if noInduk == mhs[posisi]['NIM']:
                    print(f"Data {noInduk} berhasil ditemukan diabsensi ke-{posisi+1} !!!")
                    break
                if posisi == len(mhs)-1 :
                    print(f"Data {noInduk} tidak ditemukan atau silahkan cek pengetikkannya !!!")
                    break
    def urutData(self, urutan=""):
        if urutan == "Ascending" :
            print("Data diurutkan Ascending")
            self.mhs = sorted(self.mhs, key=lambda d: d['NIM'])
        elif urutan == "Descending" :
            print("Data diurutkan Descending")
            self.mhs = sorted(self.mhs, key=lambda d: d['NIM'], reverse=True)
        else :
            print("Data tidak akan diurutkan !!!")

mhs = Mikroskil()

inputMhs = mhs.tambahData()
print(inputMhs)
print(next(inputMhs))
try :
    inputMhs.send("0347")
    inputMhs.send("Cindy")
    inputMhs.send("0852")
except StopIteration :
    print("Jumlah Mahasiswa =", mhs.jlh)

inputMhs = mhs.tambahData()
print(inputMhs)
print(next(inputMhs))
try :
    inputMhs.send("9027")
    inputMhs.send("Sintiya")
    inputMhs.send("0853")
except StopIteration :
    print("Jumlah Mahasiswa =", mhs.jlh)

inputMhs = mhs.tambahData()
print(inputMhs)
print(next(inputMhs))
try :
    inputMhs.send("1234")
    inputMhs.send("ABCDE")
    inputMhs.send("0812")
except StopIteration :
    print("Jumlah Mahasiswa =", mhs.jlh)

mhs.cetakData()
mhs.urutData()
mhs.cetakData()
mhs.urutData("Ascending")
mhs.cetakData()
mhs.urutData("Descending")
mhs.cetakData()

mhs.absensi("9027")
mhs.absensi("2839")
```

```python
# Latihan Task 2 :

class eCommerce :
    datas = []
    def tambahData(self) :
        kode, nama, jenis, jlh = "", "", "", ""
        cek = True
        datas = self.datas
        while cek:
            Kode = yield kode
            if(len(datas) < 1):
                cek = False
            else:
                cek = False
                for i in range(len(datas)):
                    if(datas[i]['Kode'] == Kode):
                        cek = True
                if(cek):
                    print("Kode sudah terdaftar ... !!!")
                    print("Masukkan Kode baru lagi ... !!!")
        Nama = yield nama
        Jenis = yield jenis
        Stok = yield jlh
        tmp = {"Kode" : Kode, "Nama" : Nama, "Jenis" : Jenis, "Stok" : Stok}
        datas.append(tmp)
    def cariData(self, kode) :
        datas = iter(self.datas)
        i = 0
        while True :
            try:
                data = next(datas)
                if(data['Kode'] == kode):
                    print(f"Data {kode} dengan nama {data['Jenis']} {data['Nama']} berhasil ditemukan di posisi ke-{i+1} !!!")
                    print(f"Sisa stok : {data['Stok']}")
                    if data['Stok'] > 1 :
                        print(data['Nama'], "masih tersisa banyak, tenang aja !!!")
                    else :
                        print(data['Nama'], "sudah [hampir] habis, silahkan restock !!!")
                    break
                i += 1
            except StopIteration:
                print(f"Data {kode} tidak ditemukan !!!")
                break
    def cetakData(self) :
        datas = self.datas
        i = 1
        print("Data barang-barang :")
        for data in datas :
            print(i, ".")
            print(f"Jenis Barang : {data['Jenis']}")
            print(f"Kode Barang  : {data['Kode']}")
            print(f"Nama Barang  : {data['Nama']}")
            print(f"Stok Barang  : {data['Stok']}")
            i+=1
    def urutData(self, urutan=""):
        if urutan == "Ascending" :
            print("Data diurutkan Ascending")
            self.datas = sorted(self.datas, key=lambda d: d['Jenis'])
        elif urutan == "Descending" :
            print("Data diurutkan Descending")
            self.datas = sorted(self.datas, key=lambda d: d['Jenis'], reverse=True)
        else :
            print("Data tidak akan diurutkan !!!")

brg = eCommerce()

# brg.tambahData("12345", "Antis", "Hand Sanitizer", 5)
inputBrg = brg.tambahData()
print(inputBrg)
next(inputBrg)
try :
    inputBrg.send("12345")
    inputBrg.send("Antis")
    inputBrg.send("Hand Sanitizer")
    inputBrg.send(5)
except StopIteration :
    print("Barang berhasil ditambahkan !!!")

# brg.tambahData("34567", "Michelin", "Ban Mobil", 0)
inputBrg = brg.tambahData()
print(inputBrg)
next(inputBrg)
try :
    inputBrg.send("34567")
    inputBrg.send("Michelin")
    inputBrg.send("Ban Mobil")
    inputBrg.send(0)
except StopIteration :
    print("Barang berhasil ditambahkan !!!")

# brg.tambahData("98765", "Apel Envy", "Buah-Buahan", 2)
inputBrg = brg.tambahData()
print(inputBrg)
next(inputBrg)
try :
    inputBrg.send("98765")
    inputBrg.send("Apel Envy")
    inputBrg.send("Buah-Buahan")
    inputBrg.send(2)
except StopIteration :
    print("Barang berhasil ditambahkan !!!")

# brg.tambahData("56789", "Dodo", "Botol Bayi", 1)
inputBrg = brg.tambahData()
print(inputBrg)
next(inputBrg)
try :
    inputBrg.send("56789")
    inputBrg.send("Dodo")
    inputBrg.send("Botol Bayi")
    inputBrg.send(1)
except StopIteration :
    print("Barang berhasil ditambahkan !!!")

# brg.tambahData("12345", "Clover", "Kue", 10)
inputBrg = brg.tambahData()
print(inputBrg)
next(inputBrg)
try :
    inputBrg.send("12345")
    inputBrg.send("Clover")
    inputBrg.send("Kue")
    inputBrg.send(10)
except StopIteration :
    print("Barang berhasil ditambahkan !!!")

brg.cariData("10293")
brg.cariData("34567")
brg.cariData("98765")
brg.urutData()
brg.cetakData()
brg.urutData("Descending")
brg.cetakData()
brg.urutData("Ascending")
brg.cetakData()
```
