```python
# Case Study

import abc

class syaratPerpus(metaclass=abc.ABCMeta):
    @abc.abstractproperty
    def namaBuku(self):
        pass
    @abc.abstractmethod
    def cetakData(self):
        pass

class perpus:
    def __init__(self, jlhBuku):
        self.jlhBuku = jlhBuku

class perpusJawa(perpus, syaratPerpus):
    namaBuku = ""
    def __init__(self, namaBuku, jlhBuku):
        super().__init__(jlhBuku)
        self.namaBuku = namaBuku
    def cetakData(self, jenisGamelan):
        print("Pada Perpus Jawa")
        print(f"Terdapat buku berjudul {self.namaBuku} dengan jumlah {self.jlhBuku} buku")
        print(f"Dan terdapat jenis gamelan berupa {jenisGamelan}")

class perpusBatak(perpus, syaratPerpus):
    namaBuku = ""
    def __init__(self, namaBuku, jlhBuku):
        super().__init__(jlhBuku)
        self.namaBuku = namaBuku
    def cetakData(self, jenisUlos):
        print("Pada Perpus Batak")
        print(f"Terdapat buku berjudul {self.namaBuku} dengan jumlah {self.jlhBuku} buku")
        print(f"Dan terdapat ulos yang berasal dari {jenisUlos}")

a = perpusJawa("Habis Gelap Terbit Terang", 10)
b = perpusBatak("Pedoman Upacara Batak", 5)

a.cetakData("Bonang")
print()
b.cetakData("Batak Toba")
```

```python
# Latihan Task 2

from abc import ABC, abstractmethod, abstractproperty

class syaratDaftarLes(ABC) :
    @abstractproperty
    def nama(self) :
        pass
    @abstractproperty
    def noHp(self) :
        pass
    @abstractmethod
    def cetakBiaya(self, nama, biaya) :
        print(f"Biaya les {nama}: {biaya}")

class lesPrivat :
    def __init__(self, nama, noHp):
        self.nama = nama
        self.noHp = noHp

class SD(lesPrivat, syaratDaftarLes) :
    nama = "tingkat SD"
    noHp = "+62"
    def __init__(self, nama, noHp, gender):
        super().__init__(nama, noHp)
        self.gender = gender
    def cetakBiaya(self, biaya=500000):
        return super().cetakBiaya(self.nama, "Rp. {:,.02f}".format(biaya))  # format currency

class SMP(lesPrivat, syaratDaftarLes) :
    nama = "tingkat SMP"
    noHp = "+62"
    def __init__(self, nama, noHp, umur):
        super().__init__(nama, noHp)
        self.umur = umur
    def cetakBiaya(self, biaya=800000):
        return super().cetakBiaya(self.nama, "Rp. {:,.02f}".format(biaya))  # format currency

class SMA(lesPrivat, syaratDaftarLes) :
    nama = "tingkat SMA"
    noHp = "+62"
    def __init__(self, nama, noHp, jurusan):
        super().__init__(nama, noHp)
        self.jurusan = jurusan
    def cetakBiaya(self, biaya=1200000):
        return super().cetakBiaya(self.nama, "Rp. {:,.02f}".format(biaya))  # format currency

andi = SD("Andi", "0852", "laki-laki")
budi = SMP("Budi", "0813", 14)
cindy = SMA("Cindy", "0853", "IPA")

andi.cetakBiaya()
budi.cetakBiaya()
cindy.cetakBiaya()
```

```python
# Latihan Task 1 :
 
class Mikroskil :
    datas = []
    def tambahData(self, noInduk, nama, noHp) :
        datas = self.datas
        if len(datas) == 0 :
            datas.append({'NIM':noInduk, 'Nama':nama, 'NoHp':noHp})
            print(f"Data {noInduk} ({nama}) berhasil ditambahkan !!!")
        else :
            for data in datas :
                if noInduk not in data['NIM'] :
                    datas.append({'NIM':noInduk, 'Nama':nama, 'NoHp':noHp})
                    print(f"Data {noInduk} ({nama}) berhasil ditambahkan !!!")
                    break
                else :
                    print(f"Data {noInduk} ({nama}) sudah ada !!!")
                    break
    def absensi(self) :
        datas = self.datas
        i = 1
        print("Data-data Mahasiswa :")
        for data in datas :
            print(i, ".")
            print(f"NIM   : {data['NIM']}")
            print(f"Nama  : {data['Nama']}")
            print(f"No Hp : {data['NoHp']}")
            print()
            i+=1
    def cariData(self, noInduk) :
        datas = self.datas
        if len(datas) == 0 :
            print("Data masih kosong, tidak ada yang bisa dicari")
        else :
            for posisi in range(len(datas)) :
                if noInduk == datas[posisi]['NIM'] :
                    print(f"Data {noInduk} berhasil ditemukan diabsensi ke-{posisi+1} !!!")
                    break
                if posisi == len(datas)-1 :
                    print(f"Data {noInduk} tidak ditemukan atau silahkan cek pengetikkannya !!!")
                    break
    def urutData(self, urutan=""):
        if urutan == "Ascending" :
            print("Data diurutkan Ascending")
            self.datas = sorted(self.datas, key=lambda d: d['NIM'])
        elif urutan == "Descending" :
            print("Data diurutkan Descending")
            self.datas = sorted(self.datas, key=lambda d: d['NIM'], reverse=True)
        else :
            print("Data tidak akan diurutkan !!!")

mhs = Mikroskil()
mhs.tambahData("211110347", "cindy", "0852")
mhs.tambahData("211110347", "cindy", "0852")
mhs.tambahData("211110123", "sintiya", "0812")
mhs.tambahData("211118429", "budi", "0896")
mhs.tambahData("211119381", "hanni", "0813")
mhs.tambahData("211115629", "dono", "0853")
mhs.cariData("211110347")
mhs.cariData("211110123")
mhs.cariData("211110123")
mhs.cariData("211115629")
mhs.cariData("211119381")
mhs.absensi()
mhs.urutData()
mhs.absensi()
mhs.urutData("Ascending")
mhs.absensi()
mhs.urutData("Descending")
mhs.absensi()
```

```python
# Latihan Task 2 :

class eCommerce :
    datas = []
    def tambahData(self, kode, nama, jenis, jlh) :
        datas = self.datas
        if len(datas) == 0 :
            datas.append({'Kode':kode, 'Nama':nama, 'Jenis':jenis, 'Stok':jlh})
            print(f"Data {kode} ({nama}) berhasil ditambahkan !!!")
        else :
            for data in datas :
                if kode not in data :
                    datas.append({'Kode':kode, 'Nama':nama, 'Jenis':jenis, 'Stok':jlh})
                    print(f"Data {kode} ({nama}) berhasil ditambahkan !!!")
                    break
                else :
                    print(f"Data {kode} ({nama}) sudah ada !!!")
                    break
    def cariData(self, kode) :
        datas = iter(self.datas)
        i = 0
        while True :
            try:
                data = next(datas)
                if(data['Kode'] == kode):
                    print(f"Data {kode} dengan nama {data['Jenis']} {data['Nama']} berhasil ditemukan di posisi ke-{i+1} !!!")
                    print(f"Sisa stok : {data['Stok']}")
                    if data['Stok'] > 1 :
                        print(data['Nama'], "masih tersisa banyak, tenang aja !!!")
                    else :
                        print(data['Nama'], "sudah [hampir] habis, silahkan restock !!!")
                    break
                i += 1
            except StopIteration:
                print(f"Data {kode} tidak ditemukan !!!")
                break
    def cetakData(self) :
        datas = self.datas
        i = 1
        print("Data barang-barang :")
        for data in datas :
            print(i, ".")
            print(f"Jenis Barang : {data['Jenis']}")
            print(f"Kode Barang  : {data['Kode']}")
            print(f"Nama Barang  : {data['Nama']}")
            print(f"Stok Barang  : {data['Stok']}")
            print()
            i+=1
    def urutData(self):
        self.datas = sorted(self.datas, key=lambda d: d['Jenis'])

brg = eCommerce()
brg.tambahData("12345", "Antis", "Hand Sanitizer", 5)
brg.tambahData("34567", "Michelin", "Ban Mobil", 0)
brg.tambahData("98765", "Apel Envy", "Buah-Buahan", 2)
brg.tambahData("56789", "Dodo", "Botol Bayi", 1)
brg.tambahData("12345", "Clover", "Kue", 10)
brg.cariData("10293")
brg.cariData("34567")
brg.cariData("98765")
brg.cetakData()
brg.urutData()
brg.cetakData()
```

```python
# Intro

def fungsiAB(a, b) :
    return a+b
    return a-b

def fungsiAB1(a, b) :
    yield a+b
    yield a-b

a = fungsiAB(2, 3)
print(a)
print(a)

b = fungsiAB1(2, 3)
print(b)
print(next(b))
print(next(b))
```

```python
import math

# x1, 2 = [(-b) +/- akar(b^2 -4*a*c)] / (2*a)

def rumusABC(a, b, c) :
    yield b**2 -4*a*c   # determinan
    yield 2*a    # nilai pembagi
    yield ((-b) + math.sqrt(b**2 - 4*a*c)) / (2*a)   # x1
    yield ((-b) - math.sqrt(b**2 - 4*a*c)) / (2*a)   # x2

a = rumusABC(2, 6, -56)
print(a)
print(f"Hasil Determinan : {next(a)}")
print(f"Nilai Pembagi    : {next(a)}")
print(f"Nilai x1         : {next(a)}")
print(f"Nilai x2         : {next(a)}")

b = rumusABC(0, 3, 7)
print(b)
print(f"Hasil Determinan : {next(b)}")
print(f"Nilai Pembagi    : {next(b)}")
print(f"Nilai x1         : {next(b)}")
print(f"Nilai x2         : {next(b)}")
```

```python
# Coroutine

def stokBarang() :
    jlh = 0 
    while True :
        brg = yield jlh
        jlh += brg

a = stokBarang()
print(f"Jumlah stok awal = {next(a)} barang")
print(f"Jumlah stok setelah ditambah 3 barang = {a.send(3)} barang")
print(f"Jumlah stok setelah ditambah 10 barang = {a.send(10)} barang")
print(f"Jumlah stok setelah dikurang 5 barang = {a.send(-5)} barang")
```

```python
# Case Study

class BUKU :
    perpus = []
    def inputBuku(self):
        kd = ""
        nm = ""
        jns = ""
        cek = True
        while cek:
            kodeBuku = yield kd
            if(len(self.perpus) < 1):
                print("Kode Kosong")
                cek = False
            else:
                cek = False
                for i in range(len(self.perpus)):
                    if(self.perpus[i]['kode'] == kodeBuku):
                        cek = True
                if(cek):
                    print("Kode Buku sudah terdaftar ... !!!")
                    print("Masukkan kode buku lagi ... !!!")
        namaBuku = yield nm
        jenisBuku = yield jns
        tmp = {"kode" : kodeBuku, "nama" : namaBuku, "jenis" : jenisBuku}
        self.perpus.append(tmp)
        print(self.perpus)
    def cetakBuku(self):
        for i in range(len(self.perpus)):
            print(f"Buku ke-{i+1}:")
            print(f"Kode Buku  = {self.perpus[i]['kode']}")
            print(f"Nama Buku  = {self.perpus[i]['nama']}")
            print(f"Jenis Buku = {self.perpus[i]['jenis']}")
            print()

a = BUKU()
b = a.inputBuku()
print(b)
print(next(b))
b.send("IF001")
b.send("Pemrograman Python")
b.send("Pemrograman")
b.send("IF002")
b.send("Bahasa C")
b.send("Pemrograman")
```
